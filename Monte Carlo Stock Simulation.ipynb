```python
# Monte Carlo Simulation of Stock Price using Geometric Brownian Motion (GBM)

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Set Parameters
S0 = 100       # Initial stock price
mu = 0.07      # Expected return (drift)
sigma = 0.2    # Volatility
T = 1          # Time to maturity in years
dt = 1/252     # Daily time step
N = int(T / dt)           # Number of time steps (252 trading days)
n_simulations = 1000      # Number of simulated paths
```

```python
# Step 2: Create a matrix to hold price paths
np.random.seed(42)  # For reproducibility
price_paths = np.zeros((N, n_simulations))
price_paths[0] = S0  # Set the first row to the initial stock price
```

```python
# Step 3: Generate random price paths
for t in range(1, N):
    z = np.random.standard_normal(n_simulations)  # Random standard normals
    price_paths[t] = price_paths[t - 1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)
```

```python
# Step 4: Plot the first 50 simulated paths
plt.figure(figsize=(12, 6))
plt.plot(price_paths[:, :50], lw=1, alpha=0.7)
plt.title('Monte Carlo Simulation: Stock Price Paths')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.grid(True)
plt.tight_layout()
plt.savefig("simulation_plot.png")  # Save image for README
plt.show()
```

```python
# Step 5: Show final stats
final_prices = price_paths[-1]
print(f"Expected Final Price: ${np.mean(final_prices):.2f}")
print(f"Standard Deviation: ${np.std(final_prices):.2f}")
```
